```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cor_df = pd.read_pickle('coronene_rse.pkl')
```

```python
sns.set_style('white')
sns.set_context('talk')
sns.set_style('ticks')
sns.boxplot(data=cor_df, palette='Dark2', linewidth=2)
plt.grid()
plt.ylim(0.01, 0.08)
plt.ylabel('RSE value')
plt.xlabel('Model type')
plt.show()
plt.savefig('coronene_box.svg')
```
